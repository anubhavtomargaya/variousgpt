```python
### JINA
import requests
def get_web_content(url):
    jina_prefit = "https://r.jina.ai/"
    query_url = f"{jina_prefit}{url}"
    response = requests.get(query_url)

    return response

def get_search_results(text):
    jina_prefit = "https://s.jina.ai/ "
    query_url = f"{jina_prefit}{text}"
    response = requests.get(query_url)

    return response
```
