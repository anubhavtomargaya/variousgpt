```python
import pandas as pd
def parse_company_filings(html_content):
    # Create a BeautifulSoup object to parse the HTML
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # Find the table in the company filings section
    table = soup.find('table', {'id': 'companyFilings'})
    rows = []
    
    # Iterate through table rows
    for row in table.find_all('tr'):
        cells = row.find_all('td')
        if cells:  # Ensure the row contains data cells
            filing_type = cells[0].get_text(strip=True)
            links = []
            years = []
            
            # Collect links and years from each cell
            for cell in cells[1:]:
                link_tag = cell.find('a')
                if link_tag:
                    year = cell.get_text(strip=True)  # Get year or disabled text
                    link = link_tag['href']  # Extract link
                    links.append(link)
                    years.append(year)
                else:
                    # If there is no link, append None
                    links.append(None)
                    year = cell.get_text(strip=True)
                    years.append(year)
            
            # Append the collected data for the current filing type
            for year, link in zip(years, links):
                rows.append({
                    "Filing Type": filing_type,
                    "Year": year,
                    "Link": link
                })

    # Create a DataFrame from the collected rows
    df = pd.DataFrame(rows)
    
    return df
```
