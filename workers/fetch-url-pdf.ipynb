```python
import json 
import requests
import pandas as pd
```

```python
def open_tij():
    with open('tij_ids.json','r') as fs:
        return json.load(fs)
```

```python

def check_id_procd(id):
    existing_repo = open_tij()
    procd = existing_repo['scraped_ids']
    return id in [int(x) for x in procd]
```

```python


def check_urls(rg_st=4000,rg_end=4400):
    # Lists to store the results
    ids = []
    quarters = []
    fys = []
    statuses = []

    # Loop through all combinations
    for id in range(rg_st,rg_end):
        if check_id_procd(id):
            print(f"{id} scraped alrdy")
        else:

            for quarter in ["Jun", "Mar"]:
                for fy in ["24", "23"]:
                    url = f"https://stockdiscovery.s3.amazonaws.com/insight/india/{id}/Conference%20Call/CC-{quarter}{fy}.pdf"
                    try:
                        print("getting url")
                        print(url)
                        response = requests.get(url)
                        if response.status_code == 200:
                            status = True
                        else:
                            status = False
                    except requests.RequestException:
                        status = False

                    # Append results to lists
                    ids.append(id)
                    quarters.append(quarter)
                    fys.append(fy)
                    statuses.append(status)

    # Build DataFrame from lists
    df = pd.DataFrame({
        'id': ids,
        'quarter': quarters,
        'fy': fys,
        'status': statuses
    })

    return df

# Example usage
df = check_urls()
print(df)
```

```python
def update_tij_repo(scraped_addn=[],present_addn=[]):
    existing_repo = open_tij()
    e_sc = existing_repo['scraped_ids']
    e_sc.extend(scraped_addn)
    e_ps = existing_repo['present_ids']
    e_ps.extend(present_addn)
    # e_ps = existing_repo['present_ids'].extend(present_addn)
    up_sm = {
        'scraped_ids':list(set(e_sc)),
        'present_ids':list(set(e_ps)),
             
             }
    with open('tij_ids.json','w') as fs:
        json.dump(up_sm,fs)

        return up_sm

    
```

```python
new_exsting_ids = list(tij_pdf_true['id'].astype('str').unique())
new_scraped_ids = list(df['id'].astype('str').unique())
```

```python
update_tij_repo(scraped_addn=new_scraped_ids,
                present_addn=new_exsting_ids)
```
