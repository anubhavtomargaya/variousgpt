```python
import requests
import pandas as pd

def check_urls():
    # Lists to store the results
    ids = []
    quarters = []
    fys = []
    statuses = []

    # Loop through all combinations
    for id in range(1, 10):
        for quarter in ["Jun", "Mar"]:
            for fy in ["24", "23"]:
                url = f"https://stockdiscovery.s3.amazonaws.com/insight/india/{id}/Conference%20Call/CC-{quarter}{fy}.pdf"
                try:
                    print("getting url")
                    print(url)
                    response = requests.get(url)
                    if response.status_code == 200:
                        status = True
                    else:
                        status = False
                except requests.RequestException:
                    status = False

                # Append results to lists
                ids.append(id)
                quarters.append(quarter)
                fys.append(fy)
                statuses.append(status)

    # Build DataFrame from lists
    df = pd.DataFrame({
        'id': ids,
        'quarter': quarters,
        'fy': fys,
        'status': statuses
    })

    return df

# Example usage
df = check_urls()
print(df)
```
